Importing all the required libraries for Beutifulsoup and then for the coversion to the dataframe

```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

Creating the dataframe from the website

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
soup = BeautifulSoup(source.content,'lxml')
table = soup.find_all('table') 
df = pd.read_html(str(table))[0]
df
```

Removing the rows where the borough is not assigned

```python
df1=df.loc[df['Borough']!='Not assigned']
df2=df1.reset_index()
postal_df=df2.drop(columns=['index'])
postal_df.head(20)
```
